```python
Transfers_Market_filename = 'Transfers.csv'
FIFA_Player_Stats_filename = 'players _ fifa20.csv'

df_playersTM = pd.read_csv(Transfers_Market_filename).loc[:,['PlayerID', 'Player', 'TeamJoined2']]
df_playersFIFA = pd.read_csv(FIFA_Player_Stats_filename).loc[:,['sofifa_id', 'short_name', 'club']]
```

## Convert DataFrame columns to a list of tuples (Player ID, Player Name, Current Team)

```python
TM = list(df_playersTM.itertuples(index = False, name = None))[:25]
```

```python
TM = list(df_playersTM.itertuples(index = False, name = None))
FIFA = list(df_playersFIFA.itertuples(index = False, name = None))
```

## String similarity (using Levenshtein algorithm) between the two lists

```python
match_list = []
start_time = time.time()
levenshtein = Levenshtein()

for tm_id, tm_player, tm_team in TM:
    
    dmax = 100
    
    for fifa_id, fifa_player, fifa_team in FIFA:

        d = levenshtein.distance(tm_player + ' - ' + tm_team, fifa_player + ' - ' + fifa_team)

        if d < dmax:
            match_row = [tm_id, tm_player, tm_team, fifa_id, fifa_player, fifa_team, d]

            if d == 0.0:
                break
            else:
                dmax = d
            
    match_list.append(match_row)

df = pd.DataFrame(match_list)    
print('Process completed in:', time.time() - start_time, 'seconds.')
```
